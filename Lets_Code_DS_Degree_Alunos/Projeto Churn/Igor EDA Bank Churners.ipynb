```python
df2=pd.get_dummies(df, columns=['Acc_status'], drop_first=False)
df2
```

```python
df.groupby('Acc_status')["Total_Relationship_Count"].mean()
```

```python
df.groupby('Acc_status')["Avg_Utilization_Ratio"].mean()
```

```python
df2['ChurnTR']=df2.groupby(['Total_Relationship_Count'])['Acc_status_Attrited Customer'].transform('sum')/df2.groupby(['Total_Relationship_Count'])['Acc_status_Existing Customer'].transform('sum')
```

```python
df2.groupby(['Total_Relationship_Count'])['ChurnTR'].mean()
```

```python
df2[['UR']]=pd.cut(df2.Avg_Utilization_Ratio, 6)
```

```python
df2['ChurnUR']=df2.groupby(['UR'])['Acc_status_Attrited Customer'].transform('sum')/df2.groupby(['UR'])['Acc_status_Existing Customer'].transform('sum')
df2.groupby(['UR'])['ChurnUR'].mean()
```

```python
df2.groupby(['Total_Relationship_Count'])[['Acc_status_Attrited Customer']].sum()/df2.groupby(['Total_Relationship_Count'])[['Acc_status_Attrited Customer']].count()
```

```python
df2.groupby(pd.qcut(df2.Months_on_book, 6))[['Acc_status_Attrited Customer']].sum()/df2.groupby(pd.qcut(df2.Months_on_book, 6))[['Acc_status_Attrited Customer']].count()
```

```python
df2.groupby(pd.qcut(df2.Avg_Utilization_Ratio, 4))[['Acc_status_Attrited Customer']].sum()/df2.groupby(pd.qcut(df2.Avg_Utilization_Ratio, 4))[['Acc_status_Attrited Customer']].count()
```
